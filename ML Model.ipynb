```python
class TradingStrategy:
    def __init__(self, name, description):
        self.name = name
        self.description = description
        self.signals = []
    
    def generate_signals(self, df):
        """Override this method in each strategy"""
        pass

class RSIMACDStrategy(TradingStrategy):
    def __init__(self, rsi_oversold=30, rsi_overbought=70, macd_threshold=0):
        super().__init__("RSI_MACD", f"RSI({rsi_oversold}/{rsi_overbought}) + MACD")
        self.rsi_oversold = rsi_oversold
        self.rsi_overbought = rsi_overbought
        self.macd_threshold = macd_threshold
    
    def generate_signals(self, df):
        signals = pd.Series(0, index=df.index)
        
        if '1h_rsi' in df.columns and '1h_macd' in df.columns:
            # Buy signals: RSI oversold + MACD positive momentum
            buy_condition = (
                (df['1h_rsi'] < self.rsi_oversold) & 
                (df['1h_macd'] > self.macd_threshold) &
                (df['1h_macd'] > df['1h_macd'].shift(1))  # MACD rising
            )
            
            # Sell signals: RSI overbought + MACD negative momentum
            sell_condition = (
                (df['1h_rsi'] > self.rsi_overbought) & 
                (df['1h_macd'] < self.macd_threshold) &
                (df['1h_macd'] < df['1h_macd'].shift(1))  # MACD falling
            )
            
            signals[buy_condition] = 1
            signals[sell_condition] = -1
        
        return signals

class TrendFollowingStrategy(TradingStrategy):
    def __init__(self, trend_threshold=0.6, momentum_periods=5):
        super().__init__("Trend_Following", f"MTF Trend + Momentum({momentum_periods})")
        self.trend_threshold = trend_threshold
        self.momentum_periods = momentum_periods
    
    def generate_signals(self, df):
        signals = pd.Series(0, index=df.index)
        
        # Calculate moving averages for trend
        sma_20 = df['1h_Close'].rolling(20).mean()
        sma_50 = df['1h_Close'].rolling(50).mean()
        
        # Trend conditions
        strong_uptrend = (
            (df['1h_Close'] > sma_20) & 
            (sma_20 > sma_50) &
            (df['1h_Close'] > df['1h_Close'].shift(self.momentum_periods))
        )
        
        strong_downtrend = (
            (df['1h_Close'] < sma_20) & 
            (sma_20 < sma_50) &
            (df['1h_Close'] < df['1h_Close'].shift(self.momentum_periods))
        )
        
        signals[strong_uptrend] = 1
        signals[strong_downtrend] = -1
        
        return signals

class VolatilityBreakoutStrategy(TradingStrategy):
    def __init__(self, volatility_threshold=1.5, volume_threshold=1.2):
        super().__init__("Volatility_Breakout", f"Vol({volatility_threshold}) + Volume({volume_threshold})")
        self.volatility_threshold = volatility_threshold
        self.volume_threshold = volume_threshold
    
    def generate_signals(self, df):
        signals = pd.Series(0, index=df.index)
        
        if '1h_atr_14' in df.columns and '1h_Volume' in df.columns:
            # Volatility and volume conditions
            avg_atr = df['1h_atr_14'].rolling(20).mean()
            avg_volume = df['1h_Volume'].rolling(20).mean()
            
            high_volatility = df['1h_atr_14'] > avg_atr * self.volatility_threshold
            high_volume = df['1h_Volume'] > avg_volume * self.volume_threshold
            
            # Breakout conditions
            resistance = df['1h_High'].rolling(20).max()
            support = df['1h_Low'].rolling(20).min()
            
            # Buy: Breakout above resistance with volume
            buy_condition = (
                high_volatility & high_volume &
                (df['1h_Close'] > resistance.shift(1))
            )
            
            # Sell: Breakdown below support with volume
            sell_condition = (
                high_volatility & high_volume &
                (df['1h_Close'] < support.shift(1))
            )
            
            signals[buy_condition] = 1
            signals[sell_condition] = -1
        
        return signals

print("✅ Trading strategy classes defined!")
```

```python
class MLEnsembleStrategy(TradingStrategy):
    def __init__(self, confidence_threshold=0.7):
        super().__init__("ML_Ensemble", f"ML Ensemble (conf>{confidence_threshold})")
        self.confidence_threshold = confidence_threshold
        self.model = None
        self.scaler = StandardScaler()
        self.feature_columns = []
    
    def prepare_features(self, df):
        """Prepare features for ML model"""
        # Select numerical features only
        feature_cols = [col for col in df.columns if 
                       not col.startswith('target_') and 
                       col not in ['1h_Open', '1h_High', '1h_Low'] and 
                       df[col].dtype in ['float64', 'int64']]
        
        # Limit to most important features to avoid overfitting
        self.feature_columns = feature_cols[:15]
        
        print(f"   Using {len(self.feature_columns)} features for ML model")
        return self.feature_columns
    
    def train_model(self, df):
        """Train the ML ensemble model"""
        print("   🤖 Training ML ensemble model...")
        
        # Prepare features
        self.prepare_features(df)
        
        if len(self.feature_columns) < 3:
            print("   ❌ Insufficient features for ML training")
            return False
        
        # Prepare data
        X = df[self.feature_columns].fillna(method='ffill').fillna(0)
        y = df['target_direction_1h'].fillna(0)
        
        # Remove rows where target is NaN
        valid_mask = ~y.isna()
        X = X[valid_mask]
        y = y[valid_mask]
        
        if len(X) < 500:
            print(f"   ⚠️ Only {len(X)} samples available for training")
            if len(X) < 100:
                return False
        
        # Split data chronologically (80% train, 20% test)
        split_idx = int(len(X) * 0.8)
        X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
        y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]
        
        # Scale features
        X_train_scaled = self.scaler.fit_transform(X_train)
        X_test_scaled = self.scaler.transform(X_test)
        
        # Create ensemble model
        rf1 = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
        rf2 = RandomForestClassifier(n_estimators=150, max_depth=8, random_state=123)
        rf3 = RandomForestClassifier(n_estimators=75, max_depth=12, random_state=456)
        
        self.model = VotingClassifier(
            estimators=[('rf1', rf1), ('rf2', rf2), ('rf3', rf3)],
            voting='soft'
        )
        
        # Train model
        self.model.fit(X_train_scaled, y_train)
        
        # Evaluate
        train_score = self.model.score(X_train_scaled, y_train)
        test_score = self.model.score(X_test_scaled, y_test)
        
        print(f"   ✅ ML Model trained - Train Acc: {train_score:.3f}, Test Acc: {test_score:.3f}")
        
        # Feature importance
        if hasattr(self.model.estimators_[0], 'feature_importances_'):
            importance = self.model.estimators_[0].feature_importances_
            top_features = sorted(zip(self.feature_columns, importance), key=lambda x: x[1], reverse=True)[:5]
            print("   📊 Top 5 features:")
            for feat, imp in top_features:
                print(f"      {feat}: {imp:.3f}")
        
        return True
    
    def generate_signals(self, df):
        signals = pd.Series(0, index=df.index)
        
        if self.model is None:
            if not self.train_model(df):
                return signals
        
        # Generate predictions
        X = df[self.feature_columns].fillna(method='ffill').fillna(0)
        
        try:
            X_scaled = self.scaler.transform(X)
            probabilities = self.model.predict_proba(X_scaled)
            
            # Get predictions and confidence
            predictions = self.model.predict(X_scaled)
            confidence = np.max(probabilities, axis=1)
            
            # Generate signals only when confidence is high
            high_confidence_mask = confidence > self.confidence_threshold
            
            # Buy signals (predict up with high confidence)
            buy_mask = (predictions == 1) & high_confidence_mask
            signals[buy_mask] = 1
            
            # Sell signals (predict down with high confidence)
            sell_mask = (predictions == 0) & high_confidence_mask
            signals[sell_mask] = -1
            
            signal_count = (signals != 0).sum()
            avg_confidence = confidence[high_confidence_mask].mean() if signal_count > 0 else 0
            print(f"   📊 Generated {signal_count} high-confidence signals (avg conf: {avg_confidence:.3f})")
            
        except Exception as e:
            print(f"   ❌ Error in ML prediction: {e}")
        
        return signals

print("✅ ML Strategy class defined!")
```

```python
class AdvancedBacktester:
    def __init__(self, initial_capital=10000, risk_per_trade=0.02, risk_reward_ratio=2.0):
        self.initial_capital = initial_capital
        self.current_capital = initial_capital
        self.risk_per_trade = risk_per_trade  # 2% risk per trade
        self.risk_reward_ratio = risk_reward_ratio  # 1:2 risk reward
        
        # Trading records
        self.trades = []
        self.equity_curve = []
        self.current_position = None
        
        # Performance metrics
        self.total_trades = 0
        self.winning_trades = 0
        self.losing_trades = 0
        self.total_profit = 0
        self.max_drawdown = 0
        self.peak_equity = initial_capital
    
    def calculate_position_size(self, entry_price, stop_loss_price, current_capital):
        """Calculate position size based on risk management"""
        if stop_loss_price == 0 or entry_price == stop_loss_price:
            return 0
        
        risk_amount = current_capital * self.risk_per_trade
        price_difference = abs(entry_price - stop_loss_price)
        
        # Position size calculation
        position_size = risk_amount / price_difference
        
        # Cap position size (max 20% of capital exposure)
        max_position_value = current_capital * 0.2
        max_position_size = max_position_value / entry_price
        
        return min(position_size, max_position_size)
    
    def calculate_stop_loss(self, entry_price, signal, atr_value, price_data):
        """Calculate dynamic stop loss based on ATR"""
        atr_multiplier = 2.0  # 2x ATR for stop loss
        
        if signal == 1:  # Buy
            stop_loss = entry_price - (atr_value * atr_multiplier)
        else:  # Sell
            stop_loss = entry_price + (atr_value * atr_multiplier)
        
        return stop_loss
    
    def calculate_take_profit(self, entry_price, stop_loss_price, signal):
        """Calculate take profit based on 1:2 risk-reward ratio"""
        risk_distance = abs(entry_price - stop_loss_price)
        reward_distance = risk_distance * self.risk_reward_ratio
        
        if signal == 1:  # Buy
            take_profit = entry_price + reward_distance
        else:  # Sell
            take_profit = entry_price - reward_distance
        
        return take_profit
    
    def enter_trade(self, timestamp, signal, entry_price, atr_value, price_data):
        """Enter a new trade"""
        if self.current_position is not None:
            return  # Already in a position
        
        # Calculate stop loss and take profit
        stop_loss = self.calculate_stop_loss(entry_price, signal, atr_value, price_data)
        take_profit = self.calculate_take_profit(entry_price, stop_loss, signal)
        
        # Calculate position size
        position_size = self.calculate_position_size(entry_price, stop_loss, self.current_capital)
        
        if position_size <= 0:
            return  # Invalid position size
        
        # Create trade record
        self.current_position = {
            'entry_time': timestamp,
            'signal': signal,
            'entry_price': entry_price,
            'stop_loss': stop_loss,
            'take_profit': take_profit,
            'position_size': position_size,
            'risk_amount': abs(entry_price - stop_loss) * position_size
        }
    
    def check_exit_conditions(self, timestamp, current_price, high_price, low_price):
        """Check if current position should be closed"""
        if self.current_position is None:
            return False
        
        position = self.current_position
        
        if position['signal'] == 1:  # Long position
            # Check stop loss (use low of candle)
            if low_price <= position['stop_loss']:
                self.close_trade(timestamp, position['stop_loss'], 'Stop Loss')
                return True
            # Check take profit (use high of candle)
            elif high_price >= position['take_profit']:
                self.close_trade(timestamp, position['take_profit'], 'Take Profit')
                return True
        
        else:  # Short position
            # Check stop loss (use high of candle)
            if high_price >= position['stop_loss']:
                self.close_trade(timestamp, position['stop_loss'], 'Stop Loss')
                return True
            # Check take profit (use low of candle)
            elif low_price <= position['take_profit']:
                self.close_trade(timestamp, position['take_profit'], 'Take Profit')
                return True
        
        return False
    
    def close_trade(self, timestamp, exit_price, exit_reason):
        """Close current position and record trade"""
        if self.current_position is None:
            return
        
        position = self.current_position
        
        # Calculate P&L
        if position['signal'] == 1:  # Long
            price_diff = exit_price - position['entry_price']
        else:  # Short
            price_diff = position['entry_price'] - exit_price
        
        profit = price_diff * position['position_size']
        profit_pct = (profit / self.current_capital) * 100
        
        # Update capital
        self.current_capital += profit
        self.total_profit += profit
        
        # Record trade
        trade_record = {
            'entry_time': position['entry_time'],
            'exit_time': timestamp,
            'signal': 'BUY' if position['signal'] == 1 else 'SELL',
            'entry_price': position['entry_price'],
            'exit_price': exit_price,
            'stop_loss': position['stop_loss'],
            'take_profit': position['take_profit'],
            'position_size': position['position_size'],
            'profit': profit,
            'profit_pct': profit_pct,
            'exit_reason': exit_reason,
            'duration_hours': (timestamp - position['entry_time']).total_seconds() / 3600,
            'risk_reward_actual': abs(profit / position['risk_amount']) if position['risk_amount'] != 0 else 0
        }
        
        self.trades.append(trade_record)
        
        # Update statistics
        self.total_trades += 1
        if profit > 0:
            self.winning_trades += 1
        else:
            self.losing_trades += 1
        
        # Update drawdown
        if self.current_capital > self.peak_equity:
            self.peak_equity = self.current_capital
        
        current_drawdown = (self.peak_equity - self.current_capital) / self.peak_equity
        self.max_drawdown = max(self.max_drawdown, current_drawdown)
        
        # Clear position
        self.current_position = None

print("✅ Advanced Backtesting Engine defined!")
```

```python
def run_backtest(self, df, signals, strategy_name="Strategy"):
    """Run complete backtest"""
    print(f"🔄 Running backtest for {strategy_name}...")
    
    # Reset state
    self.__init__(self.initial_capital, self.risk_per_trade, self.risk_reward_ratio)
    
    # Ensure we have required columns
    required_cols = ['1h_Open', '1h_High', '1h_Low', '1h_Close']
    missing_cols = [col for col in required_cols if col not in df.columns]
    if missing_cols:
        print(f"❌ Missing required columns: {missing_cols}")
        return self.get_performance_metrics()
    
    # Get ATR for stop loss calculation
    atr_column = '1h_atr_14' if '1h_atr_14' in df.columns else None
    
    for i, (timestamp, row) in enumerate(df.iterrows()):
        # Record equity
        self.equity_curve.append({
            'timestamp': timestamp,
            'equity': self.current_capital,
            'position': 1 if self.current_position else 0
        })
        
        # Check exit conditions first
        if self.current_position:
            self.check_exit_conditions(
                timestamp, 
                row['1h_Close'], 
                row['1h_High'], 
                row['1h_Low']
            )
        
        # Check for new entry signals
        if self.current_position is None and timestamp in signals.index:
            signal = signals.loc[timestamp]
            
            if signal != 0:  # Valid signal
                # Use current close as entry price (realistic execution)
                entry_price = row['1h_Close']
                
                # Get ATR value for stop loss
                atr_value = row[atr_column] if atr_column else (row['1h_High'] - row['1h_Low'])
                atr_value = max(atr_value, 0.1)  # Minimum ATR
                
                self.enter_trade(timestamp, signal, entry_price, atr_value, df.loc[:timestamp])
    
    # Close any remaining position at the end
    if self.current_position:
        last_price = df['1h_Close'].iloc[-1]
        self.close_trade(df.index[-1], last_price, 'End of Data')
    
    print(f"✅ Backtest completed: {self.total_trades} trades")
    return self.get_performance_metrics()

def get_performance_metrics(self):
    """Calculate comprehensive performance metrics"""
    if self.total_trades == 0:
        return {
            'total_trades': 0, 'win_rate': 0, 'total_return': 0,
            'max_drawdown': 0, 'profit_factor': 0, 'avg_win': 0,
            'avg_loss': 0, 'sharpe_ratio': 0, 'final_capital': self.current_capital
        }
    
    # Basic metrics
    win_rate = self.winning_trades / self.total_trades
    total_return = (self.current_capital - self.initial_capital) / self.initial_capital
    
    # Profit/Loss analysis
    trades_df = pd.DataFrame(self.trades)
    
    if len(trades_df) > 0:
        winning_trades = trades_df[trades_df['profit'] > 0]
        losing_trades = trades_df[trades_df['profit'] < 0]
        
        avg_win = winning_trades['profit'].mean() if len(winning_trades) > 0 else 0
        avg_loss = losing_trades['profit'].mean() if len(losing_trades) > 0 else 0
        
        total_wins = winning_trades['profit'].sum() if len(winning_trades) > 0 else 0
        total_losses = abs(losing_trades['profit'].sum()) if len(losing_trades) > 0 else 0
        
        profit_factor = total_wins / total_losses if total_losses > 0 else float('inf')
        
        # Risk-adjusted returns
        if len(self.equity_curve) > 1:
            equity_df = pd.DataFrame(self.equity_curve)
            equity_returns = equity_df['equity'].pct_change().dropna()
            
            if len(equity_returns) > 0 and equity_returns.std() > 0:
                sharpe_ratio = equity_returns.mean() / equity_returns.std() * np.sqrt(365 * 24)
            else:
                sharpe_ratio = 0
        else:
            sharpe_ratio = 0
    else:
        avg_win = avg_loss = profit_factor = sharpe_ratio = 0
    
    return {
        'total_trades': self.total_trades,
        'winning_trades': self.winning_trades,
        'losing_trades': self.losing_trades,
        'win_rate': win_rate,
        'total_return': total_return,
        'total_return_pct': total_return * 100,
        'max_drawdown': self.max_drawdown,
        'max_drawdown_pct': self.max_drawdown * 100,
        'profit_factor': profit_factor,
        'avg_win': avg_win,
        'avg_loss': avg_loss,
        'sharpe_ratio': sharpe_ratio,
        'final_capital': self.current_capital,
        'total_profit': self.total_profit
    }

# Add methods to AdvancedBacktester class
AdvancedBacktester.run_backtest = run_backtest
AdvancedBacktester.get_performance_metrics = get_performance_metrics

print("✅ Backtesting execution methods added!")
```

```python
class HighWinRateBacktester(AdvancedBacktester):
    def __init__(self, initial_capital=10000, risk_per_trade=0.01, risk_reward_ratio=2.0):
        # Lower risk per trade (1%) for higher win rate strategies
        super().__init__(initial_capital, risk_per_trade, risk_reward_ratio)
        self.min_win_rate_target = 0.75  # 75% target
    
    def calculate_stop_loss(self, entry_price, signal, atr_value, price_data):
        """Tighter stop loss for high win rate strategies"""
        atr_multiplier = 1.5  # Tighter stop loss (1.5x ATR instead of 2x)
        
        if signal == 1:  # Buy
            stop_loss = entry_price - (atr_value * atr_multiplier)
        else:  # Sell
            stop_loss = entry_price + (atr_value * atr_multiplier)
        
        return stop_loss
    
    def enter_trade(self, timestamp, signal, entry_price, atr_value, price_data):
        """Enhanced entry with additional validations"""
        if self.current_position is not None:
            return
        
        # Additional validation before entering trade
        if not self.validate_entry_conditions(timestamp, signal, entry_price, price_data):
            return
        
        # Calculate stop loss and take profit
        stop_loss = self.calculate_stop_loss(entry_price, signal, atr_value, price_data)
        take_profit = self.calculate_take_profit(entry_price, stop_loss, signal)
        
        # Calculate position size
        position_size = self.calculate_position_size(entry_price, stop_loss, self.current_capital)
        
        if position_size <= 0:
            return
        
        # Create trade record
        self.current_position = {
            'entry_time': timestamp,
            'signal': signal,
            'entry_price': entry_price,
            'stop_loss': stop_loss,
            'take_profit': take_profit,
            'position_size': position_size,
            'risk_amount': abs(entry_price - stop_loss) * position_size
        }
    
    def validate_entry_conditions(self, timestamp, signal, entry_price, price_data):
        """Additional validation for high-quality entries"""
        
        # Get recent data
        recent_data = price_data.tail(10)
        if len(recent_data) < 5:
            return False
        
        # 1. Avoid trading during low liquidity periods
        if timestamp.hour in [22, 23, 0, 1, 2, 3, 4, 5]:  # Asian session
            return False
        
        # 2. Check if market is in consolidation (avoid whipsaws)
        recent_range = recent_data['1h_High'].max() - recent_data['1h_Low'].min()
        avg_range = recent_data['1h_atr_14'].iloc[-1] * 10 if '1h_atr_14' in recent_data.columns else recent_range
        
        if recent_range < avg_range * 0.3:  # Too low volatility
            return False
        
        # 3. Avoid trading immediately after big moves
        if len(recent_data) >= 3:
            last_3_moves = recent_data['1h_Close'].iloc[-3:].pct_change().abs()
            if last_3_moves.max() > 0.01:  # Avoid if recent move > 1%
                return False
        
        return True

class OptimizedBacktester(HighWinRateBacktester):
    def __init__(self, initial_capital=10000):
        super().__init__(initial_capital, risk_per_trade=0.015, risk_reward_ratio=2.0)
        # Track additional metrics for optimization
        self.consecutive_losses = 0
        self.max_consecutive_losses = 3
        self.daily_trades = {}
        self.max_daily_trades = 2
    
    def enter_trade(self, timestamp, signal, entry_price, atr_value, price_data):
        """Enhanced trade entry with additional filters"""
        
        # Check consecutive losses limit
        if self.consecutive_losses >= self.max_consecutive_losses:
            return  # Stop trading after max consecutive losses
        
        # Check daily trade limit
        trade_date = timestamp.date()
        daily_count = self.daily_trades.get(trade_date, 0)
        if daily_count >= self.max_daily_trades:
            return  # Max trades per day reached
        
        # Call parent method
        super().enter_trade(timestamp, signal, entry_price, atr_value, price_data)
        
        # Update daily trade count
        if self.current_position:
            self.daily_trades[trade_date] = daily_count + 1
    
    def close_trade(self, timestamp, exit_price, exit_reason):
        """Track consecutive losses"""
        if self.current_position is None:
            return
        
        # Calculate profit before closing
        position = self.current_position
        if position['signal'] == 1:
            price_diff = exit_price - position['entry_price']
        else:
            price_diff = position['entry_price'] - exit_price
        
        profit = price_diff * position['position_size']
        
        # Update consecutive losses
        if profit <= 0:
            self.consecutive_losses += 1
        else:
            self.consecutive_losses = 0  # Reset on winning trade
        
        # Call parent method
        super().close_trade(timestamp, exit_price, exit_reason)

print("✅ Enhanced backtester for high win rate strategies created!")
```

```python
class UltraConservativeStrategy(HighWinRateStrategy):
    """Ultra-conservative strategy targeting 80%+ win rate"""
    
    def __init__(self):
        super().__init__("Ultra_Conservative", min_confluence=5)
        self.cooldown_periods = 12  # Wait 12 hours between trades
        self.last_trade_time = None
    
    def generate_signals(self, df):
        signals = pd.Series(0, index=df.index)
        
        for i in range(100, len(df)):  # Need more history
            current_time = df.index[i]
            
            # Cooldown check
            if self.last_trade_time and (current_time - self.last_trade_time).total_seconds() < self.cooldown_periods * 3600:
                continue
            
            row = df.iloc[i]
            confluences = self.check_confluence(df, i)
            
            # ULTRA STRICT criteria for entry
            # Buy criteria
            if (len(confluences['buy']) >= 5 and 
                '1h_rsi' in df.columns and row['1h_rsi'] < 25 and  # Very oversold
                self.check_support_confluence(df, i) and
                self.check_volume_surge(df, i) and
                self.check_trend_reversal_pattern(df, i, 'buy')):
                
                signals.iloc[i] = 1
                self.last_trade_time = current_time
                
            # Sell criteria  
            elif (len(confluences['sell']) >= 5 and
                  '1h_rsi' in df.columns and row['1h_rsi'] > 75 and  # Very overbought
                  self.check_resistance_confluence(df, i) and
                  self.check_volume_surge(df, i) and
                  self.check_trend_reversal_pattern(df, i, 'sell')):
                
                signals.iloc[i] = -1
                self.last_trade_time = current_time
        
        return signals
    
    def check_support_confluence(self, df, idx):
        """Check for strong support confluence"""
        if 'distance_to_support' not in df.columns:
            return False
        
        row = df.iloc[idx]
        # Very close to support (within 5 pips)
        return row['distance_to_support'] < 5
    
    def check_resistance_confluence(self, df, idx):
        """Check for strong resistance confluence"""
        if 'distance_to_resistance' not in df.columns:
            return False
        
        row = df.iloc[idx]
        # Very close to resistance (within 5 pips)
        return row['distance_to_resistance'] < 5
    
    def check_volume_surge(self, df, idx):
        """Check for volume surge (confirmation)"""
        if '1h_Volume' not in df.columns or '1h_volume_sma_20' not in df.columns:
            return True  # Skip check if no volume data
        
        row = df.iloc[idx]
        # Volume must be at least 50% above average
        return row['1h_Volume'] > row['1h_volume_sma_20'] * 1.5
    
    def check_trend_reversal_pattern(self, df, idx, direction):
        """Check for trend reversal patterns"""
        if idx < 10:
            return False
        
        recent_closes = df['1h_Close'].iloc[idx-5:idx+1]
        
        if direction == 'buy':
            # Look for higher low pattern
            return recent_closes.iloc[-1] > recent_closes.iloc[-3]
        else:  # sell
            # Look for lower high pattern
            return recent_closes.iloc[-1] < recent_closes.iloc[-3]

class MeanReversionMasterStrategy(HighWinRateStrategy):
    """Mean reversion strategy with multiple timeframe confirmation"""
    
    def __init__(self):
        super().__init__("Mean_Reversion_Master", min_confluence=4)
    
    def generate_signals(self, df):
        signals = pd.Series(0, index=df.index)
        
        for i in range(50, len(df)):
            if self.is_extreme_oversold(df, i):
                if self.confirm_reversal_signals(df, i, 'buy'):
                    signals.iloc[i] = 1
            
            elif self.is_extreme_overbought(df, i):
                if self.confirm_reversal_signals(df, i, 'sell'):
                    signals.iloc[i] = -1
        
        return signals
    
    def is_extreme_oversold(self, df, idx):
        """Check for extreme oversold conditions"""
        if '1h_rsi' not in df.columns:
            return False
        
        row = df.iloc[idx]
        
        # Multiple oversold confirmations
        conditions = [
            row['1h_rsi'] < 20,  # Very oversold RSI
        ]
        
        # Add Stochastic if available
        if '1h_stoch_k' in df.columns:
            conditions.append(row['1h_stoch_k'] < 20)
        
        # Add Bollinger Band position if available
        if '1h_bb_position' in df.columns:
            conditions.append(row['1h_bb_position'] < 0.1)  # Below lower BB
        
        return sum(conditions) >= 2  # At least 2 confirmations
    
    def is_extreme_overbought(self, df, idx):
        """Check for extreme overbought conditions"""
        if '1h_rsi' not in df.columns:
            return False
        
        row = df.iloc[idx]
        
        # Multiple overbought confirmations
        conditions = [
            row['1h_rsi'] > 80,  # Very overbought RSI
        ]
        
        # Add Stochastic if available
        if '1h_stoch_k' in df.columns:
            conditions.append(row['1h_stoch_k'] > 80)
        
        # Add Bollinger Band position if available
        if '1h_bb_position' in df.columns:
            conditions.append(row['1h_bb_position'] > 0.9)  # Above upper BB
        
        return sum(conditions) >= 2  # At least 2 confirmations
    
    def confirm_reversal_signals(self, df, idx, direction):
        """Confirm reversal with additional signals"""
        confirmations = 0
        
        # Volume confirmation
        if ('1h_Volume' in df.columns and '1h_volume_sma_20' in df.columns and
            df.iloc[idx]['1h_Volume'] > df.iloc[idx]['1h_volume_sma_20'] * 1.2):
            confirmations += 1
        
        # Support/Resistance confirmation
        if direction == 'buy' and 'distance_to_support' in df.columns:
            if df.iloc[idx]['distance_to_support'] < 10:
                confirmations += 1
        elif direction == 'sell' and 'distance_to_resistance' in df.columns:
            if df.iloc[idx]['distance_to_resistance'] < 10:
                confirmations += 1
        
        # Divergence confirmation (simplified)
        if '1h_macd' in df.columns and idx >= 5:
            macd_current = df['1h_macd'].iloc[idx]
            macd_prev = df['1h_macd'].iloc[idx-5]
            price_current = df['1h_Close'].iloc[idx]
            price_prev = df['1h_Close'].iloc[idx-5]
            
            if direction == 'buy':
                # Bullish divergence: price lower but MACD higher
                if price_current < price_prev and macd_current > macd_prev:
                    confirmations += 1
            else:
                # Bearish divergence: price higher but MACD lower
                if price_current > price_prev and macd_current < macd_prev:
                    confirmations += 1
        
        return confirmations >= 2

# Test ultra-conservative strategies
print("🎯 Testing Ultra-Conservative Strategies...")

ultra_strategies = [
    UltraConservativeStrategy(),
    MeanReversionMasterStrategy()
]

ultra_results = []

for strategy in ultra_strategies:
    print(f"\n📊 Testing: {strategy.name}")
    
    signals = strategy.generate_signals(df)
    signal_count = (signals != 0).sum()
    
    print(f"   📡 Generated {signal_count} signals")
    
    if signal_count > 0:
        backtester = OptimizedBacktester(initial_capital=10000)
        performance = backtester.run_backtest(df, signals, strategy.name)
        performance['strategy'] = strategy.name
        performance['signal_count'] = signal_count
        
        ultra_results.append(performance)
        
        print(f"   📈 Return: {performance['total_return_pct']:.2f}%")
        print(f"   🎯 Win Rate: {performance['win_rate']:.1%}")
        print(f"   💰 Profit Factor: {performance['profit_factor']:.2f}")
        print(f"   📊 Trades: {performance['total_trades']}")
        
        if performance['win_rate'] >= 0.75:
            print("   🏆 TARGET ACHIEVED! ✅")

# Combine all results
all_high_wr_results = high_wr_results + ultra_results

print(f"\n🎯 SUMMARY: Tested {len(all_high_wr_results)} high win rate strategies")
```

```python
# Practical High Win Rate Strategies - Less Conservative, More Signals
print("🔧 CREATING PRACTICAL HIGH WIN RATE STRATEGIES")
print("="*80)

class PracticalHighWinStrategy(TradingStrategy):
    def __init__(self, name, description):
        super().__init__(name, description)
    
    def check_basic_conditions(self, df, idx):
        """Check basic market conditions"""
        if idx < 20:  # Need some history
            return False
        
        row = df.iloc[idx]
        
        # Basic data availability check
        required_cols = ['1h_Close', '1h_High', '1h_Low', '1h_Open']
        for col in required_cols:
            if col not in df.columns or pd.isna(row[col]):
                return False
        
        return True

class RSIExtremesStrategy(PracticalHighWinStrategy):
    def __init__(self, rsi_oversold=25, rsi_overbought=75):
        super().__init__("RSI_Extremes", f"RSI Extremes ({rsi_oversold}/{rsi_overbought}) + Confirmation")
        self.rsi_oversold = rsi_oversold
        self.rsi_overbought = rsi_overbought
    
    def generate_signals(self, df):
        signals = pd.Series(0, index=df.index)
        
        if '1h_rsi' not in df.columns:
            print("   ⚠️ RSI column not found, calculating RSI...")
            # Calculate RSI if not present
            delta = df['1h_Close'].diff()
            gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
            loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
            rs = gain / loss
            df['1h_rsi'] = 100 - (100 / (1 + rs))
        
        for i in range(20, len(df)):
            if not self.check_basic_conditions(df, i):
                continue
            
            row = df.iloc[i]
            
            # Buy when RSI is oversold
            if row['1h_rsi'] < self.rsi_oversold:
                # Additional confirmation: price not in free fall
                recent_closes = df['1h_Close'].iloc[i-5:i+1]
                if recent_closes.iloc[-1] >= recent_closes.iloc[-3]:  # Some stability
                    signals.iloc[i] = 1
            
            # Sell when RSI is overbought
            elif row['1h_rsi'] > self.rsi_overbought:
                # Additional confirmation: price not in strong rally
                recent_closes = df['1h_Close'].iloc[i-5:i+1]
                if recent_closes.iloc[-1] <= recent_closes.iloc[-3]:  # Some weakness
                    signals.iloc[i] = -1
        
        return signals

class BollingerBounceStrategy(PracticalHighWinStrategy):
    def __init__(self):
        super().__init__("Bollinger_Bounce", "Bollinger Band Bounce + Volume")
    
    def generate_signals(self, df):
        signals = pd.Series(0, index=df.index)
        
        # Calculate Bollinger Bands if not present
        if '1h_bb_upper' not in df.columns:
            print("   ⚠️ Bollinger Bands not found, calculating...")
            bb_period = 20
            bb_std = 2
            sma = df['1h_Close'].rolling(window=bb_period).mean()
            std = df['1h_Close'].rolling(window=bb_period).std()
            df['1h_bb_upper'] = sma + (std * bb_std)
            df['1h_bb_lower'] = sma - (std * bb_std)
            df['1h_bb_middle'] = sma
        
        for i in range(25, len(df)):
            if not self.check_basic_conditions(df, i):
                continue
            
            row = df.iloc[i]
            
            # Buy when price bounces off lower Bollinger Band
            if (row['1h_Close'] <= row['1h_bb_lower'] * 1.005 and  # Near or below lower BB
                row['1h_Close'] > row['1h_Low']):  # Not the lowest point of candle
                
                # Volume confirmation if available
                volume_ok = True
                if '1h_Volume' in df.columns and '1h_volume_sma_20' in df.columns:
                    volume_ok = row['1h_Volume'] > row['1h_volume_sma_20'] * 0.8
                
                if volume_ok:
                    signals.iloc[i] = 1
            
            # Sell when price bounces off upper Bollinger Band
            elif (row['1h_Close'] >= row['1h_bb_upper'] * 0.995 and  # Near or above upper BB
                  row['1h_Close'] < row['1h_High']):  # Not the highest point of candle
                
                # Volume confirmation if available
                volume_ok = True
                if '1h_Volume' in df.columns and '1h_volume_sma_20' in df.columns:
                    volume_ok = row['1h_Volume'] > row['1h_volume_sma_20'] * 0.8
                
                if volume_ok:
                    signals.iloc[i] = -1
        
        return signals

class SupportResistanceBounceStrategy(PracticalHighWinStrategy):
    def __init__(self):
        super().__init__("Support_Resistance_Bounce", "S/R Level Bounces")
    
    def calculate_support_resistance(self, df, lookback=20):
        """Calculate dynamic support and resistance levels"""
        support_levels = df['1h_Low'].rolling(window=lookback).min()
        resistance_levels = df['1h_High'].rolling(window=lookback).max()
        return support_levels, resistance_levels
    
    def generate_signals(self, df):
        signals = pd.Series(0, index=df.index)
        
        # Calculate support and resistance
        support_levels, resistance_levels = self.calculate_support_resistance(df)
        
        for i in range(25, len(df)):
            if not self.check_basic_conditions(df, i):
                continue
            
            row = df.iloc[i]
            support = support_levels.iloc[i]
            resistance = resistance_levels.iloc[i]
            
            # Buy when price bounces off support
            if (row['1h_Low'] <= support * 1.003 and  # Touched support
                row['1h_Close'] > row['1h_Low'] and  # Bounced (closed above low)
                row['1h_Close'] > support):  # Closed above support
                
                signals.iloc[i] = 1
            
            # Sell when price bounces off resistance
            elif (row['1h_High'] >= resistance * 0.997 and  # Touched resistance
                  row['1h_Close'] < row['1h_High'] and  # Rejected (closed below high)
                  row['1h_Close'] < resistance):  # Closed below resistance
                
                signals.iloc[i] = -1
        
        return signals

class MomentumReversalStrategy(PracticalHighWinStrategy):
    def __init__(self):
        super().__init__("Momentum_Reversal", "Momentum Exhaustion Reversal")
    
    def generate_signals(self, df):
        signals = pd.Series(0, index=df.index)
        
        for i in range(15, len(df)):
            if not self.check_basic_conditions(df, i):
                continue
            
            # Look for momentum exhaustion
            recent_closes = df['1h_Close'].iloc[i-10:i+1]
            recent_highs = df['1h_High'].iloc[i-10:i+1]
            recent_lows = df['1h_Low'].iloc[i-10:i+1]
            
            current_close = recent_closes.iloc[-1]
            
            # Bullish reversal: after downtrend, small green candle
            if (recent_closes.iloc[0] > recent_closes.iloc[-3] and  # Overall decline
                current_close > df['1h_Open'].iloc[i] and  # Current candle is green
                current_close < recent_closes.iloc[-2]):  # But not making new highs
                
                # Additional filter: not at extreme low
                if current_close > recent_lows.min() * 1.002:
                    signals.iloc[i] = 1
            
            # Bearish reversal: after uptrend, small red candle  
            elif (recent_closes.iloc[0] < recent_closes.iloc[-3] and  # Overall rise
                  current_close < df['1h_Open'].iloc[i] and  # Current candle is red
                  current_close > recent_closes.iloc[-2]):  # But not making new lows
                
                # Additional filter: not at extreme high
                if current_close < recent_highs.max() * 0.998:
                    signals.iloc[i] = -1
        
        return signals

class SimpleTrendPullbackStrategy(PracticalHighWinStrategy):
    def __init__(self):
        super().__init__("Simple_Trend_Pullback", "Simple Moving Average Pullback")
    
    def generate_signals(self, df):
        signals = pd.Series(0, index=df.index)
        
        # Calculate simple moving averages
        if '1h_sma_20' not in df.columns:
            df['1h_sma_20'] = df['1h_Close'].rolling(window=20).mean()
        
        for i in range(25, len(df)):
            if not self.check_basic_conditions(df, i):
                continue
            
            row = df.iloc[i]
            sma_20 = row['1h_sma_20']
            
            # Only trade if we have valid SMA
            if pd.isna(sma_20):
                continue
            
            # Determine trend using SMA slope
            sma_current = sma_20
            sma_prev = df['1h_sma_20'].iloc[i-5]
            
            # Uptrend: buy on pullback to SMA
            if sma_current > sma_prev * 1.001:  # SMA rising
                if (row['1h_Close'] <= sma_20 * 1.003 and  # Near SMA
                    row['1h_Close'] > sma_20 * 0.997):  # But not too far below
                    signals.iloc[i] = 1
            
            # Downtrend: sell on pullback to SMA
            elif sma_current < sma_prev * 0.999:  # SMA falling
                if (row['1h_Close'] >= sma_20 * 0.997 and  # Near SMA
                    row['1h_Close'] < sma_20 * 1.003):  # But not too far above
                    signals.iloc[i] = -1
        
        return signals

print("✅ Practical high win rate strategies created!")
```

```python
def test_practical_strategies(df):
    """Test practical high win rate strategies"""
    print("\n🎯 TESTING PRACTICAL HIGH WIN RATE STRATEGIES")
    print("="*80)
    
    # Create practical strategies
    strategies = [
        RSIExtremesStrategy(rsi_oversold=20, rsi_overbought=80),
        RSIExtremesStrategy(rsi_oversold=25, rsi_overbought=75),
        RSIExtremesStrategy(rsi_oversold=30, rsi_overbought=70),
        BollingerBounceStrategy(),
        SupportResistanceBounceStrategy(),
        MomentumReversalStrategy(),
        SimpleTrendPullbackStrategy()
    ]
    
    results = []
    
    for i, strategy in enumerate(strategies):
        print(f"\n📊 Testing Strategy {i+1}/{len(strategies)}: {strategy.name}")
        print(f"   📝 {strategy.description}")
        
        try:
            # Generate signals
            signals = strategy.generate_signals(df)
            signal_count = (signals != 0).sum()
            buy_signals = (signals == 1).sum()
            sell_signals = (signals == -1).sum()
            
            print(f"   📡 Total Signals: {signal_count} (Buy: {buy_signals}, Sell: {sell_signals})")
            
            if signal_count == 0:
                print("   ⚠️ No signals generated")
                continue
            
            # Run backtest with standard parameters
            backtester = AdvancedBacktester(
                initial_capital=10000,
                risk_per_trade=0.02,  # 2% risk
                risk_reward_ratio=2.0  # 1:2 ratio
            )
            
            performance = backtester.run_backtest(df, signals, strategy.name)
            performance['strategy'] = strategy.name
            performance['description'] = strategy.description
            performance['signal_count'] = signal_count
            performance['buy_signals'] = buy_signals
            performance['sell_signals'] = sell_signals
            performance['trades_df'] = pd.DataFrame(backtester.trades) if backtester.trades else pd.DataFrame()
            
            results.append(performance)
            
            # Print detailed results
            print(f"   📈 Total Return: {performance['total_return_pct']:.2f}%")
            print(f"   🎯 Win Rate: {performance['win_rate']:.1%}")
            print(f"   💰 Profit Factor: {performance['profit_factor']:.2f}")
            print(f"   📉 Max Drawdown: {performance['max_drawdown_pct']:.2f}%")
            print(f"   📊 Total Trades: {performance['total_trades']}")
            print(f"   ⚡ Sharpe Ratio: {performance['sharpe_ratio']:.2f}")
            
            # Check if target achieved
            if performance['win_rate'] >= 0.75:
                print("   🏆 ACHIEVED 75%+ WIN RATE TARGET! ✅")
            elif performance['win_rate'] >= 0.65:
                print("   🥈 Good win rate (65%+) 👍")
            elif performance['win_rate'] >= 0.55:
                print("   🥉 Decent win rate (55%+) 👌")
            else:
                print("   📊 Below 55% win rate")
            
            # Additional analysis
            if not performance['trades_df'].empty:
                trades = performance['trades_df']
                avg_duration = trades['duration_hours'].mean()
                print(f"   ⏰ Avg Trade Duration: {avg_duration:.1f} hours")
                
                if len(trades) >= 5:
                    win_trades = trades[trades['profit'] > 0]
                    lose_trades = trades[trades['profit'] <= 0]
                    print(f"   💚 Wins: {len(win_trades)} (Avg: ${win_trades['profit'].mean():.2f})")
                    print(f"   💔 Losses: {len(lose_trades)} (Avg: ${lose_trades['profit'].mean():.2f})")
            
        except Exception as e:
            print(f"   ❌ Error testing strategy: {str(e)}")
            continue
    
    return results

# Test practical strategies
practical_results = test_practical_strategies(df)

# Display summary
if practical_results:
    print(f"\n📊 PRACTICAL STRATEGIES SUMMARY")
    print("="*60)
    
    # Sort by win rate
    sorted_results = sorted(practical_results, key=lambda x: x['win_rate'], reverse=True)
    
    print("🏆 TOP STRATEGIES BY WIN RATE:")
    for i, result in enumerate(sorted_results[:5], 1):
        print(f"{i}. {result['strategy']}")
        print(f"   🎯 Win Rate: {result['win_rate']:.1%}")
        print(f"   📈 Return: {result['total_return_pct']:.2f}%")
        print(f"   💰 Profit Factor: {result['profit_factor']:.2f}")
        print(f"   📊 Trades: {result['total_trades']}")
        print()
    
    # Check if any achieved target
    high_win_strategies = [r for r in sorted_results if r['win_rate'] >= 0.75]
    if high_win_strategies:
        print(f"🎉 SUCCESS! {len(high_win_strategies)} strategies achieved 75%+ win rate!")
        
        best_strategy = high_win_strategies[0]
        print(f"\n🏆 BEST HIGH WIN RATE STRATEGY: {best_strategy['strategy']}")
        print(f"   🎯 Win Rate: {best_strategy['win_rate']:.1%}")
        print(f"   📈 Total Return: {best_strategy['total_return_pct']:.2f}%")
        print(f"   💰 Profit Factor: {best_strategy['profit_factor']:.2f}")
        print(f"   📉 Max Drawdown: {best_strategy['max_drawdown_pct']:.2f}%")
        print(f"   📊 Total Trades: {best_strategy['total_trades']}")
    else:
        print("📊 No strategies achieved 75% win rate yet. Adjusting parameters...")
else:
    print("❌ No successful strategies found")
```

```python
def analyze_best_strategy(results_df, detailed_results):
    """Detailed analysis of the best performing strategy"""
    if results_df is None or results_df.empty:
        print("❌ No strategies to analyze")
        return
    
    print("\n🏆 DETAILED ANALYSIS OF BEST STRATEGY")
    print("="*60)
    
    # Get best strategy
    best_strategy_name = results_df.iloc[0]['Strategy']
    best_result = next(r for r in detailed_results if r['strategy'] == best_strategy_name)
    
    print(f"🥇 CHAMPION STRATEGY: {best_strategy_name}")
    print(f"📝 Description: {results_df.iloc[0]['Description']}")
    print("="*60)
    
    # Performance Summary
    print("📊 PERFORMANCE SUMMARY:")
    print(f"   💰 Initial Capital: $10,000")
    print(f"   🎯 Final Capital: ${best_result['final_capital']:,.2f}")
    print(f"   📈 Total Return: {best_result['total_return_pct']:.2f}%")
    print(f"   📅 Period: {len(df)} hours ({len(df)/24:.0f} days)")
    print(f"   📊 Annualized Return: {(best_result['total_return_pct'] * 365) / (len(df)/24):.1f}%")
    
    print(f"\n🎯 TRADING STATISTICS:")
    print(f"   🔢 Total Trades: {best_result['total_trades']}")
    print(f"   ✅ Winning Trades: {best_result['winning_trades']} ({best_result['win_rate']:.1%})")
    print(f"   ❌ Losing Trades: {best_result['losing_trades']}")
    print(f"   💎 Profit Factor: {best_result['profit_factor']:.2f}")
    print(f"   📈 Average Win: ${best_result['avg_win']:.2f}")
    print(f"   📉 Average Loss: ${best_result['avg_loss']:.2f}")
    
    print(f"\n⚖️ RISK METRICS:")
    print(f"   📉 Maximum Drawdown: {best_result['max_drawdown_pct']:.2f}%")
    print(f"   ⚡ Sharpe Ratio: {best_result['sharpe_ratio']:.2f}")
    print(f"   🎯 Risk per Trade: 2.00%")
    print(f"   💰 Risk-Reward Ratio: 1:2")
    
    # Trade Analysis
    if not best_result['trades_df'].empty:
        trades_df = best_result['trades_df']
        
        print(f"\n📊 TRADE ANALYSIS:")
        print(f"   ⏰ Average Trade Duration: {trades_df['duration_hours'].mean():.1f} hours")
        print(f"   🏆 Best Trade: ${trades_df['profit'].max():.2f}")
        print(f"   💥 Worst Trade: ${trades_df['profit'].min():.2f}")
        print(f"   📈 Win Streak: {calculate_max_streak(trades_df, 'win')}")
        print(f"   📉 Loss Streak: {calculate_max_streak(trades_df, 'loss')}")
        
        # Monthly performance
        trades_df['exit_time'] = pd.to_datetime(trades_df['exit_time'])
        monthly_profits = trades_df.groupby(trades_df['exit_time'].dt.to_period('M'))['profit'].sum()
        winning_months = (monthly_profits > 0).sum()
        total_months = len(monthly_profits)
        
        print(f"   📅 Profitable Months: {winning_months}/{total_months} ({winning_months/total_months:.1%})")
        print(f"   📊 Best Month: ${monthly_profits.max():.2f}")
        print(f"   📉 Worst Month: ${monthly_profits.min():.2f}")
    
    # Risk Assessment
    print(f"\n🚨 RISK ASSESSMENT:")
    risk_score = assess_strategy_risk(best_result)
    print(f"   🎯 Overall Risk Score: {risk_score}/10")
    
    if best_result['max_drawdown_pct'] < 5:
        print("   ✅ Low Risk: Drawdown < 5%")
    elif best_result['max_drawdown_pct'] < 10:
        print("   ⚠️ Medium Risk: Drawdown 5-10%")
    else:
        print("   🚨 High Risk: Drawdown > 10%")
    
    if best_result['win_rate'] > 0.6:
        print("   ✅ High Win Rate: >60%")
    elif best_result['win_rate'] > 0.4:
        print("   ⚠️ Moderate Win Rate: 40-60%")
    else:
        print("   🚨 Low Win Rate: <40%")
    
    print(f"\n💡 STRATEGY INSIGHTS:")
    print(f"   🎯 This strategy excels at: {get_strategy_strengths(best_result)}")
    print(f"   ⚠️ Watch out for: {get_strategy_weaknesses(best_result)}")
    print(f"   📈 Best market conditions: {get_optimal_conditions(best_strategy_name)}")
    
    return best_result

def calculate_max_streak(trades_df, streak_type):
    """Calculate maximum winning or losing streak"""
    if trades_df.empty:
        return 0
    
    if streak_type == 'win':
        wins = (trades_df['profit'] > 0).astype(int)
    else:
        wins = (trades_df['profit'] <= 0).astype(int)
    
    max_streak = 0
    current_streak = 0
    
    for win in wins:
        if win:
            current_streak += 1
            max_streak = max(max_streak, current_streak)
        else:
            current_streak = 0
    
    return max_streak

def assess_strategy_risk(result):
    """Assess overall strategy risk (1-10 scale)"""
    score = 10
    
    # Penalize high drawdown
    if result['max_drawdown_pct'] > 15:
        score -= 4
    elif result['max_drawdown_pct'] > 10:
        score -= 2
    elif result['max_drawdown_pct'] > 5:
        score -= 1
    
    # Penalize low win rate
    if result['win_rate'] < 0.3:
        score -= 3
    elif result['win_rate'] < 0.4:
        score -= 2
    elif result['win_rate'] < 0.5:
        score -= 1
    
    # Penalize low profit factor
    if result['profit_factor'] < 1.0:
        score -= 3
    elif result['profit_factor'] < 1.2:
        score -= 1
    
    # Penalize few trades
    if result['total_trades'] < 10:
        score -= 2
    elif result['total_trades'] < 20:
        score -= 1
    
    return max(1, score)

def get_strategy_strengths(result):
    """Identify strategy strengths"""
    strengths = []
    
    if result['win_rate'] > 0.6:
        strengths.append("high win rate")
    if result['profit_factor'] > 2.0:
        strengths.append("excellent profit factor")
    if result['max_drawdown_pct'] < 5:
        strengths.append("low drawdown")
    if result['sharpe_ratio'] > 1.0:
        strengths.append("good risk-adjusted returns")
    
    return ", ".join(strengths) if strengths else "consistent performance"

def get_strategy_weaknesses(result):
    """Identify strategy weaknesses"""
    weaknesses = []
    
    if result['win_rate'] < 0.4:
        weaknesses.append("low win rate")
    if result['max_drawdown_pct'] > 10:
        weaknesses.append("high drawdown periods")
    if result['total_trades'] < 20:
        weaknesses.append("limited trade sample")
    if result['profit_factor'] < 1.2:
        weaknesses.append("low profit factor")
    
    return ", ".join(weaknesses) if weaknesses else "extended losing streaks"

def get_optimal_conditions(strategy_name):
    """Get optimal market conditions for strategy"""
    conditions = {
        'RSI_MACD': "oversold/overbought reversals with momentum confirmation",
        'Trend_Following': "strong trending markets with clear direction",
        'Volatility_Breakout': "high volatility periods with volume surges",
        'ML_Ensemble': "diverse market conditions with clear patterns"
    }
    
    for key in conditions:
        if key in strategy_name:
            return conditions[key]
    
    return "various market conditions"

# Analyze best strategy
if results_df is not None:
    best_strategy_analysis = analyze_best_strategy(results_df, detailed_results)
```
